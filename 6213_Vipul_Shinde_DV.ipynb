```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
restaurents = df['City'].value_counts()
plt.figure(figsize=(10, 5))
restaurents.plot(kind = 'bar' , color = 'blue')
plt.xlabel('Cities')
plt.ylabel('Count')
plt.title('Total Restaurents')
plt.show()
```

**Resultant Observation from above graph**

This chart displays the distribution of restaurants across various cities, with the number of restaurants in each city indicated by the height of the blue bars.

```python
plt.figure(figsize=(18,9))
sns.countplot(data=df,x='Avg ratings')
plt.xticks(rotation=70)
plt.xlabel('Avg Ratings', weight='bold',fontsize=30)
plt.ylabel('Count', weight='bold',fontsize=30)
plt.show()
```


**Resultant Observation from above graph**

This countplot illustrates the distribution of average ratings for a dataset, showing how many data points fall into each average rating category while also addressing the issue of crowded x-axis labels by rotating them for better readability.

```python
df["Area"].value_counts().head(20).plot(kind="bar", xlabel="Area", ylabel="Count", title="top 20 areas of restaurants")
plt.show()
```

**Resultant Observation from above graph**

This bar chart visually represents the 20 areas with the most restaurants, offering insights into the distribution of restaurants across different areas.

```python
plt.scatter(df["Avg ratings"], df["City"])
plt.show()
```

**Resultant Observation from above graph**

This scatter plot displays the distribution of average ratings across different cities, allowing for an examination of any potential patterns or outliers in the data.

```python
df["Avg ratings"].plot(kind="hist")
plt.show()
```

**Resultant Observation from above graph**

This histogram provides an overview of the frequency distribution of average ratings, showing the concentration of ratings within various value ranges.

```python
plt.figure(figsize=(16,9))
ax = sns.countplot(data=df, x='City',palette='Blues')
for label in ax.containers:
    ax.bar_label(label)
plt.show()
```

**Resultant Observation from above graph**

This countplot visually represents the distribution of data points (e.g., restaurants) across different cities, with labels indicating the exact count for each city.

```python
cities = ['Pune', 'Surat', 'Ahmedabad', 'Chennai', 'Kolkata', 'Mumbai', 'Hyderabad', 'Bangalore']
avg_price = [353, 270, 318, 356, 362, 392, 299, 382]

sns.barplot(x=cities, y=avg_price,color='#367be5')

plt.xlabel('City')
plt.ylabel('Average Price')
plt.title('City Food Expenses Bar Plot')

plt.xticks(rotation=45)

sns.set_style('whitegrid')

plt.grid(True, alpha=0.5)

plt.show()
```

**Resultant Observation from above graph**

This bar plot visually compares the average food expenses in different cities, highlighting variations in food costs across the specified cities.

```python
city = df.City.value_counts()

plt.bar(city.index,city.values,color='red')

plt.title("Restaurant counts by city",fontdict={'fontweight':'bold'})
plt.xticks(rotation='vertical')
plt.xlabel("Cities")
plt.ylabel("Counts")
for i,v in enumerate(city.values):
    plt.text(i,v,str(v),ha='center')
plt.show()
```

**Resultant Observation from above graph**

This bar chart provides a clear representation of restaurant counts in various cities, enabling easy comparison of the number of restaurants in each city.

```python
# 1. ID
print('ID Column:')
print('Number of unique IDs:',df['ID'].nunique())
print('Minimum ID:',df['ID'].min())
print('Maximum ID:',df['ID'].max())
print('')

# 2. Area
print('Area Column:')
print('Number of unique Areas:',df['Area'].nunique())
print('')

# 3. City
print('City Column:')
print('Number of unique Cities:',df['City'].nunique())
print('Count of each City:',df['City'].value_counts())
print('')

# 4. Restaurant
print('Restaurant Column:')
print('Number of unique Restaurants:',df['Restaurant'].nunique())
print('Count of each Restaurants:',df['Restaurant'].value_counts())
print('')

# 5. Price
print('Price Column:')
print('Minimum Price:',df['Price'].min())
print('Maximum Price:',df['Price'].max())
print('Mean Price:',df['Price'].mean())
print('Median Price:',df['Price'].median())
print('')

# 6. Avg ratings
print('Avg ratings:')
print('Minimum Avg ratings:',df['Avg ratings'].min())
print('Maximum Avg ratings:',df['Avg ratings'].max())
print('Mean Avg ratings:',df['Avg ratings'].mean())
print('Median Avg ratings:',df['Avg ratings'].median())
print('')

# 7. Total ratings
print('Total ratings:')
print('Minimum Total ratings:',df['Total ratings'].min())
print('Maximum Total ratings:',df['Total ratings'].max())
print('Mean Total ratings:',df['Total ratings'].mean())
print('Median Total ratings:',df['Total ratings'].median())
print('')

# 8. Food type
print('Food type Column:')
print('Number of unique Food type:',df['Food type'].nunique())
print('Count of each Food Type:',df['Food type'].value_counts())
print('')

# 9. Address
print('Address Column:')
print('Number of unique Address:',df['Address'].nunique())
print('Count of each Address:',df['Address'].value_counts())
print('')

# 10. Delivery time
print('Delivery time Column:')
print('Minimum Delivery time:',df['Delivery time'].min())
print('Maximum Delivery time:',df['Delivery time'].max())
print('Mean Delivery time:',df['Delivery time'].mean())
print('Median Delivery time:',df['Delivery time'].median())
print('')

# Example: creating a histogram for Avg ratings
plt.hist(df['Avg ratings'],bins = 20,edgecolor = 'k')
plt.xlabel('Avg ratings')
plt.ylabel('Frequency')
plt.title('Avg rating distribution')
plt.show()
```
